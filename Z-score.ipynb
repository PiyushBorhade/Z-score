#  Removing outliers by using Z-score techique!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
df
```

```python
df['Height'].mean()
```

```python
df['Height'].std()
```

```python
(df['Height'] - df['Height'].mean()) / df['Height'].std()
```

```python
df['zscore'] = (df['Height'] - df['Height'].mean())/df['Height'].std()
df
```

```python
df[df['zscore'] > 3]
```

```python
df[df['zscore'] < -3]
```

```python
out_df = df[(df['zscore'] > 3) | (df['zscore'] < -3)]
out_df
```

```python
no_out = df[(df['zscore'] < 3) & (df['zscore'] > -3)]
no_out
```

**Here we can observe that the outliers are removed by using Z-score techique!**
